```python
import random

turno = random.randint(1, 2)
print(f"turno : jugador {turno}")
jugador1 = 0
jugador2 = 0

while True:
    input("presione ENTER para seguir jugando")
    dado = random.randint(1, 6)
    print(f"dado : {dado}")

    if turno == 1 :
         jugador1 += dado
         print(f"jugador1 : {jugador1}")

         if jugador1 == 3:
            print("caiste en una escalera, sube a la casilla : 22")
            jugador1 = 22
         elif jugador1 == 5:

            print("caiste en una escalera, sube a la casilla : 8")
            jugador1 = 8
         elif jugador1 == 20:
            print("caiste en una escalera, sube a la casilla : 29")
            jugador1 = 29
         elif jugador1 == 19:
            print("caiste en una serpiente, baja a la casilla : 7")
            jugador1 = 7
         elif jugador1 == 21:
            print("caiste en una serpiente, baja a la casilla : 9")
            jugador1 = 9
         elif jugador1 == 19:
            print("caiste en una serpiente, baja a la casilla : 27")
            jugador1 = 1

            turno = 2

    else:
         jugador2 += dado
         print(f"jugador2 : {jugador2}")

         if jugador2 == 3:
            print("caiste en una escalera, sube a la casilla : 22")
            jugador2 = 22
         elif jugador2 == 5:
            print("caiste en una escalera, sube a la casilla : 8")
            jugador2 = 8
         elif jugador2 == 20:
            print("caiste en una escalera, sube a la casilla : 29")
            jugador2 = 29
         elif jugador2 == 19:
            print("caiste en una serpiente, baja a la casilla : 7")
            jugador2 = 7
         elif jugador2 == 21:
            print("caiste en una serpiente, baja a la casilla : 9")
            jugador2 = 9
         elif jugador2 == 19:
            print("caiste en una serpiente, baja a la casilla : 27")
            jugador2 = 1

            turno = 1

    if jugador1 == 30:
         break
    elif jugador1 > 30:
         jugador1 -= dado
         print(f"has caido fuera del tablero regrese a la casilla {jugador1}")
    if jugador2 == 30:
         break
    elif jugador2 > 30:
         jugador2 -= dado
         print(f"has caido fuera del tablero regrese a la casilla {jugador2}")
```
